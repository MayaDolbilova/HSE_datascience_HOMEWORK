 Create a plot with graphs of two functions: $y=e^{-x^2}$ and $y=3x^2 e^{-x^2}$. Add x and y axes labels and the legend. Turn on the grid. Customize curve colors and styles.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0,5,18)
e= np.linspace(0,5,18)
y1=e**(-x**2)
y2=3*x**2*e**(-x**2) 
ax.plot(x, y1, '-C0', label="$y=e^{-x^2}$")
ax.plot(x, y2, '-.C1', label="$y=3x^2 e^{-x^2}$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$");
ax.legend(["$y=e^{-x^2}$", "$y=3x^2 e^{-x^2}$"], loc ="upper right");
ax.grid();
```

File "apple_vs_cabot.csv" that you can find at "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/" contains stock prices of "Apple Inc." and "Cabot Oil & Gas" recorded from 08 Feb 2013 till 07 Feb 2018. For simplicity instead of the actual date a day number within this range is stored in the file. Using an example above read this file and create a line plot demonstrating variation of theses two prices.

```python
import requests
import matplotlib.pyplot as plt
import numpy as np
#getting information from a file
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"
file_name = "apple_vs_cabot.csv"
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200
#cleaning data from a file
s = web_data.text.replace("\n", "")
str_dat = s.split(",")
str_dat.pop(0)
str_dat.pop(0)
str_dat.pop(0)
dat = [float(n) for n in str_dat]
#turning list to a numpy array
b = np.array(dat)
#give it a second dimension
k=int(len(b)/2)
c= b.reshape(k,2)
#creating coordinates
x=np.arange(len(b)/2)
y1=c[:,0]
y2=c[:,1]
#making a line plot
fig, ax = plt.subplots()
ax.plot(x, y1, '-C0', label="Apple Inc")
ax.plot(x, y2, '-.C1', label="Cabot Oil & Gas")
ax.set_xlabel("days")
ax.set_ylabel("prices");
ax.legend(["Apple Inc", "Cabot Oil & Gas"], loc ="upper left");
ax.grid();
```

 Above is the dataset describing world total energy supply recorded at 2015. Create a bar plot for it. Write text labels with values on each bar. Add x and y axes labels.

```python
# Total energy supply (TES) by source, World 2015
# Unit of measurement: Thousand tons of oil equivalent (ktoe)
names = ["Coal", "Natural\n gas", "Nuclear", "Hydro", "Wind,\n solar,\n etc", "Biofuels\n and waste", "Oil"]
dat = [3842742,2928795,670172,334851,203821,1271235,4328233]
import matplotlib.pyplot as plt
import numpy as np
datt = [int(n) for n in dat]
y=datt[:]
fig, ax = plt.subplots()
ax.bar(names[0:7],y[0:7],width=0.9)
ax.set_ylabel('ktoe')
ax.set_xlabel('TES by source')

rects = ax.patches
labels = [f"{i}" for i in dat]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width()/2, height+5, label,ha="center", va="bottom"
    )

plt.show()

```

File "moscow_temp.csv" that you can find at "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/" contains daily records of average temperature in Moscow for many years. Read this file and create a histogram showing how often certain temperatures are registered. The data are represented in Fahrenheit. Before plotting convert them into Celsius. Add appropriate axes labels.

```python
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"
file_name = "moscow_temp.csv"
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200
s = web_data.text.replace("\n", ",")
str_dat = s.split(",")
str_dat.pop(-1)
data=np.array(str_dat)
print(len(data))
dat=data.reshape(int(len(data)/4),4)
p=[float(i) for i in dat[1:,3]]
celcius=[int((i-32)/1.8) for i in p]
c=sorted(celcius)
fig, ax = plt.subplots()
ax.hist(c, bins=30, edgecolor='k', rwidth=0.75)
ax.set_xlabel("Temperature(in Celsius)")
ax.set_ylabel("Frequency")
 
```
