File "bank_customers.csv" that you can find at "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/" describes bank clinets. It contains three columns: customer age (CustAge), period of relationship with the bank in months (Months), and credit limit (CredLim). Plot a histogram for the column Month and compute its mean, median and the mode. Show these values on the histogram. Make a second copy of the histogram, compute quartiles and show them on the plot.

```python
import requests
import matplotlib.pyplot as plt
from collections import Counter
def my_quantile(data, q):
    inx = int(round(q * len(data)))
    return sorted(data)[inx]

# This is an URL of a repository
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"

# We need this file
file_name = "bank_customers.csv"

# Here we downlaod the file
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200

# Take a look at the data
str_data = web_data.text.splitlines()
lst_data = [s.split(',') for s in str_data[1:]]

# Take only month
data_month = [int(s[1]) for s in lst_data]
print(data_month[:10])

#Create plots with hist
fig, ax = plt.subplots()
ax.hist(data_month, bins=max(data_month) - min(data_month)+1);
ax.set_xlabel("period of relationship with bank (month)")
ax.set_ylabel("amount of bank clients")
ax.grid();

#counting mean and add it to the hist

mean=sum(data_month) / len(data_month)
ax.axvline(mean, color='red', linewidth=3, label=f'mean at {int(mean)}')
ax.legend()
ax.grid();

#counting median and add it to the hist

srt_datamont=sorted(data_month)
size = len(srt_datamont)
median = srt_datamont[size//2]
ax.axvline(median, color='green', linewidth=3, label=f'median at {median}')
ax.legend()
ax.grid();

#counting mode and add it to the hist

bins = Counter(data_month)
mode = bins.most_common(1)[0][0]
ax.axvline(mode, color='cyan', linewidth=3, label=f'mode at {mode}')    
ax.legend()
ax.grid();


#quartiles
Q1=my_quantile(data_month, 0.25)
Q2=median
Q3=my_quantile(data_month, 0.75)

fig, ax = plt.subplots()
ax.hist(data_month, bins=max(data_month) - min(data_month)+1);
ax.set_xlabel("period of relationship with bank (month)")
ax.set_ylabel("amount of bank clients")
ax.axvline(Q1, color='C1', linewidth=3, label=f'Q1 at {Q1}')
ax.axvline(Q2, color='C2', linewidth=3, label=f'Q2 at {Q2}')
ax.axvline(Q3, color='C3', linewidth=3, label=f'Q3 at {Q3}')
ax.legend()
ax.grid();

```

File "bank_customers.csv" that you can find at "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/" describes bank clinets. It contains three columns: customer age (CustAge), period of relationship with the bank in months (Months), and credit limit (CredLim). Plot a histogram for the column CustAge and compute its standard deviation and the interquartile range. Show them on the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
import requests

def my_stddev(data):
    
    size = len(data)
    mean = sum(data) / size
    sqr_dev = [(x - mean)**2 for x in data]
    variance = sum(sqr_dev) / (size - 1)
    return variance**0.5



def my_quantile(data, q):
    inx = int(round(q * len(data)))
    return sorted(data)[inx]


# This is an URL of a repository
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"

# We need this file
file_name = "bank_customers.csv"

# Here we downlaod the file
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200

# Take a look at the data
str_data = web_data.text.splitlines()
lst_data = [s.split(',') for s in str_data[1:]]

# Take only custage 
data_custage = [int(s[0]) for s in lst_data]
print(data_custage[:10])

#build hist
fig, ax = plt.subplots()
ax.hist(data_custage, bins=max(data_custage) - min(data_custage)+1);
ax.set_xlabel("customer age")
ax.set_ylabel("amount of bank clients")
ax.grid();

#counting standart deviation
stand_deviation=my_stddev(data_custage)
mean=sum(data_custage) / len(data_custage)
ax.axvline(mean, color='red', linewidth=3, label=f'mean at {int(mean)}')
ax.axvspan(-stand_deviation + mean, stand_deviation + mean, color='yellow', alpha=0.3, label=f'$\sigma={stand_deviation:4.2f}$')
ax.legend()
ax.grid();

#interquartile range

Q1 = my_quantile(data_custage, 0.25)
Q3 = my_quantile(data_custage, 0.75)
ax.axvspan(Q1, Q3, color='green', alpha=0.8, label=f'iqr={Q3-Q1:4.2f}')
ax.legend()
ax.grid();
```

Download the dataset from the previous exercise and compute correlation coefficients for each pair of columns.

```python
import matplotlib.pyplot as plt
import numpy as np
import requests

def my_mean(data):
    return sum(data) / len(data)

def my_corrcoef(xs, ys):
    x_mean = my_mean(xs)
    y_mean = my_mean(ys)
    x_std = my_stddev(xs)
    y_std = my_stddev(ys)
    xd = [x - x_mean for x in xs]
    yd = [y - y_mean for y in ys]
    dot = [x * y for x, y in zip(xd, yd)]
    cov = sum(dot) / (len(dot) - 1)
    return cov / (x_std * y_std)



# This is an URL of a repository
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"

# We need this file
file_name = "bank_customers.csv"

# Here we downlaod the file
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200

# Take a look at the data
str_data = web_data.text.splitlines()
lst_data = [s.split(',') for s in str_data[1:]]

#getting data for all columns

data_custage = [int(s[0]) for s in lst_data]
print(data_custage[:10])

data = [s[2] for s in lst_data]
data_credlim=[]
for i in data:
    c=float(i)
    k=int(c)
    data_credlim.append(k)
print(data_credlim[:20])

data_month = [int(s[1]) for s in lst_data]
print(data_month[:10])

#counting correlation coefficients
cor_coefficient_one=my_corrcoef(data_custage, data_month)
cor_coefficient_two=my_corrcoef(data_month,data_credlim)
cor_coefficient_three=my_corrcoef(data_credlim, data_custage)

print(cor_coefficient_one)
print(cor_coefficient_two)
print(cor_coefficient_three)
```
