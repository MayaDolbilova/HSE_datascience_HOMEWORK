File "canberra_weather.csv" that you can find at "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/" contains daily weather observations for Canberra, Australian Capital Territory for June 2020. Read this file and create a line plot that show how minimal and maximal temperature changed day by day. Put labels with corresponding dates on x axes.4

```python
import numpy as np
import matplotlib.pyplot as plt
import requests

base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"
file_name = "canberra_weather.csv"

web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200

data=web_data.text[974:]
a=data.split(",")

c=np.array(a)
q=np.delete(c,0)
k=int(len(q)/21)
b=q.reshape(k,21)
fig, ax = plt.subplots()
y1=[i[1] for i in b]
ys1=list(map(float, y1))
y2=[i[2] for i in b]
ys2=list(map(float, y2))

x=[i[0] for i in b]
ax.plot(x,ys1, '-C0')
ax.plot(x,ys2,'-C1')

tc=x[1::2]
ax.set_xticks(tc) 
ax.set_xticklabels(tc, rotation=40);
ax.grid();
ax.legend(["MIN C°", "MAX C°"])
```

Create 3D surface plot of a function $z=x^2-y^2$. Add axes labels and color bar.

```python
def f (x,y):
    z=x**2-y**2
    return z
x = np.linspace(-15, 15, 101)
y = np.linspace(-5, 10, 50)
x, y = np.meshgrid(x, y)
z = f(x, y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, z, cmap='hot')  

fig.colorbar(surf, shrink=0.6, aspect=20, pad=0.15);
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
